```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df=pd.read_csv("titanic-training-data.csv")
```

```python
plt.hist(x=df["Age"],color="red")
plt.title("distributoin of Age")
plt.xlabel("Age")
plt.ylabel("frequency")
plt.show()
```

```python
plt.hist(x=df["Age"],color="red")
plt.title("distributoin of Age",fontsize=20,color="blue")
plt.xlabel("Age")
plt.ylabel("frequency")
plt.show()
```

```python
plt.boxplot(x=df["Age"])
plt.title("distributoin of Age",fontsize=20,color="blue")
plt.xlabel("Age")
plt.ylabel("frequency")
plt.show()
```

```python
sns.boxplot(x=df["Age"])
plt.title("distributoin of Age",fontsize=20,color="blue")
plt.xlabel("Age")
plt.ylabel("frequency")
plt.show()
```

```python
df["Sex"].value_counts().plot(kind="bar")
plt.title("Sex Distributoin")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
df["Sex"].value_counts().plot(kind="barh" ,color="teal")
plt.title("Sex Distributoin")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
df["Pclass"].value_counts().plot(kind="barh" ,color="teal")
plt.title("Sex Distributoin")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
from matplotlib import cm
emb= df["Embarked"].value_counts()
keys=emb.keys().to_list()
counts=emb.to_list()
cs=cm.Set1([2,4,6,8])
plt.pie(x=counts,labels=keys,autopct='%1.1f%%',colors=cs)
plt.show()
```
